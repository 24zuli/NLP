```python

example_string = """The world of possibilities is endless, and the human mind, with its boundless curiosity, often seeks to explore the uncharted realms of imagination and reality. Imagine a small village perched on the edge of a cliff overlooking a vast, shimmering ocean. The village is alive with the chatter of its inhabitants, each contributing their unique tales and traditions to the vibrant tapestry of community life. The air is filled with the aroma of freshly baked bread, mingling with the salty breeze rolling in from the sea. In the heart of this village stands an ancient tree, its gnarled roots spreading deep into the earth, and its branches reaching for the heavens. This tree, known as the Tree of Whispers, is said to hold the secrets of the past, present, and future. Every evening, the villagers gather beneath its sprawling canopy, sharing stories of their day, their hopes, and their fears. They believe the tree listens, and in return, it grants wisdom to those who truly hear its rustling leaves. Across the ocean lies an island shrouded in mist, a place of legends and mysteries. Sailors speak of treasures hidden within its dense forests, guarded by creatures both magnificent and terrifying. The villagers often gaze at the distant island, their imaginations painting vivid pictures of what might lie beyond the horizon. One day, a young girl named Lila, with eyes as bright as the morning sun, decides to set sail and uncover the truths hidden in the tales. She builds a small boat with the help of the village elder, an old man whose hands are as rough as the bark of the Tree of Whispers but whose heart is gentle and kind. As she sets out on her journey, the villagers gather to bid her farewell, their cheers echoing across the water. The ocean, vast and unpredictable, tests her courage with towering waves and fierce winds. But Lila, guided by the stars and the whispered wisdom of the tree, perseveres. Her adventure is filled with discoveries—unfamiliar creatures, breathtaking landscapes, and encounters with people whose customs and stories differ vastly from her own. Each experience enriches her understanding of the world and her place within it. When she finally reaches the misty island, she finds not only the legendary treasure but also a profound sense of connection to the earth, the sea, and the sky. Returning home, Lila shares her journey with the villagers, inspiring them to embrace curiosity and courage. The Tree of Whispers seems to rustle louder that evening, as if rejoicing in her tale. In this small village, a new tradition begins—each year, someone sets out on a journey, returning with stories that weave the fabric of their collective identity ever richer. The village thrives, not just on the exchange of goods but on the exchange of ideas and dreams, proving that the greatest treasure is often found in the stories we share and the bonds we create."""
```

# Tokenizing

```python
sent_tokenize(example_string)
```

```python
tokens = word_tokenize(example_string)
```

```python
tokens
```

# Filtering Stopwords

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
word_quote = example_string
```

```python
words_in_quote = word_tokenize(word_quote)
```

```python
words_in_quote
```

```python
stop_words = set(stopwords.words("english"))
```

```python
filtered_list = []
```

```python
for word in words_in_quote:
    if word.casefold() not in stop_words:
        filtered_list.append(word)
```

```python
filtered_list = [
    word for word in words_in_quote if word.casefold() not in stop_words
]
```

```python
filtered_list
```

# Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = PorterStemmer()
```

```python
string_for_stemming = example_string
```

```python
words = word_tokenize(string_for_stemming)
```

```python
words
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
```

```python
stemmed_words
```

# Frequncy

```python
from nltk.probability import FreqDist
```

```python
freq_dist = FreqDist(tokens)
```

```python
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
print(freq_dist.most_common(10)) 
```
